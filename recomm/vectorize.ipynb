```python
#import numpy as np
#import pandas as pd
from gensim.models import Word2Vec
import pickle
import time
import logging
logging.basicConfig(
	format='%(asctime)s : %(levelname)s : %(message)s',
	level=logging.INFO)

with open('corpus/user-based_attr.list', 'rb') as f:
    corpus_attr = pickle.load(f)
start = time.time()
model_attr = Word2Vec(corpus_attr,
                      size=300,
                      window=99999,
                      min_count=0,
                      workers=2,
                      iter=100,
                      sg=1,
                      sample=1e-1)
print("Elapsed time: %s sec" % (time.time() - start))
```
